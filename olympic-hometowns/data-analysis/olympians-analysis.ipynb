# Constants

```python
# df with state populations
pop_df = pd.read_csv("state_populations_2023.csv")
pop_df["Geographic Area"] = pop_df["Geographic Area"].apply(lambda x: x.replace(".", ""))
pop_df["2023"] = pop_df["2023"].apply(lambda x: int(x.replace(",", "")))
pop_df.head(10)
```

```python
# function to generate list of columns to remove from grouped by df
def remove_cols(df, keep):
    cols = list(df.columns)
    for col in keep:
        cols.remove(col)
    return cols
```

```python
# function to count instances in specific column
def group_by_col(df, col):
    grouped = df.groupby(col).count().sort_values(by="FIRST_NAME", ascending=False).drop(columns=remove_cols(df, [col, "FIRST_NAME"])).reset_index().rename(columns={col: col.title().replace("_", " "), "FIRST_NAME": "Athletes"})
    return grouped
```

```python
# function to count instances in specific column
def group_by_cols(df, cols):
    rename = {}
    for col in cols:
        rename[col] = col.title().replace("_", " ")
    rename["FIRST_NAME"] = "Athletes"
    grouped = df.groupby(cols).count().sort_values(by="FIRST_NAME", ascending=False).drop(columns=remove_cols(df, cols + ["FIRST_NAME"])).reset_index().rename(columns=rename)
    return grouped
```

```python
# function to count instances per state relative to state population
def state_proportional(df, col):
    compare_df = pop_df.rename(columns={"Geographic Area": col})
    pop_compare = pd.merge(compare_df, df, on=col)
    pop_compare["Athletes per 100,000 people"] = (pop_compare["Athletes"] * 100000) / pop_compare["2023"]
    return pop_compare
```

# How many people moved?

```python
origins_df["STAYED_PUT"] = origins_df.apply(lambda x: x["HOMETOWN_STATE"] == x["CURRENT_RESIDENCE_STATE"], axis=1)
stayed_put_counts = group_by_col(origins_df, "STAYED_PUT")
stayed_put_counts.sort_values(by="Athletes", ascending=False)
```

```python
states = set(origins_df["HOMETOWN_STATE"].unique() + origins_df["CURRENT_RESIDENCE_STATE"].unique())
perc_left_dict = {
    'State': [],
    'Perc. left': []
}
for state in states:
    state_df = origins_df.loc[origins_df["HOMETOWN_STATE"] == state]
    total = len(state_df.index)
    left_count = len(state_df.loc[~state_df["STAYED_PUT"]].index)
    perc_left = left_count/total*100
    perc_left_dict['State'].append(state)
    perc_left_dict['Perc. left'].append(perc_left)
    
```
